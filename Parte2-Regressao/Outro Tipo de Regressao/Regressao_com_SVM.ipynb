```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

Base Plano de Saúde

```python
base_plano_saude2 = pd.read_csv('../plano_saude2.csv')
x_plano_saude2 = base_plano_saude2.iloc[:, 0:1].values
y_plano_saude2 = base_plano_saude2.iloc[:, 1].values
x_teste_arvore = np.arange(min(x_plano_saude2), max(x_plano_saude2), 0.1)
x_teste_arvore = x_teste_arvore.reshape(-1, 1)
```

```python
x_plano_saude2
```

```python
y_plano_saude2
```

```python
# Kernel Linear
from sklearn.svm import SVR
regressor_svr_saude_linear = SVR(kernel = 'linear')
regressor_svr_saude_linear.fit(x_plano_saude2, y_plano_saude2)
```

```python
grafico = px.scatter(x= x_plano_saude2.ravel(), y= y_plano_saude2)
grafico.add_scatter(x= x_teste_arvore.ravel(), y= regressor_svr_saude_linear.predict(x_teste_arvore), name= 'Regressão')
grafico.show()
```

```python
# Kernel Polinomial
regressor_svr_saude_poly = SVR(kernel='poly', degree=4)
regressor_svr_saude_poly.fit(x_plano_saude2, y_plano_saude2)
```

```python
grafico = px.scatter(x= x_plano_saude2.ravel(), y= y_plano_saude2)
grafico.add_scatter(x= x_teste_arvore.ravel(), y= regressor_svr_saude_poly.predict(x_teste_arvore), name= 'Regressão')
grafico.show()
```

```python
# Kernel rbf
regressor_svr_saude_rbf = SVR(kernel='rbf')
regressor_svr_saude_rbf.fit(x_plano_saude2, y_plano_saude2)
```

```python
grafico = px.scatter(x= x_plano_saude2.ravel(), y= y_plano_saude2.ravel())
grafico.add_scatter(x= x_plano_saude2.ravel(), y= regressor_svr_saude_rbf.predict(x_plano_saude2), name= 'Regressão')
grafico.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
x_plano_saude2_scaled = scaler_x.fit_transform(x_plano_saude2)
scaler_y = StandardScaler()
y_plano_saude2_scaled = scaler_y.fit_transform(y_plano_saude2.reshape(-1, 1))
```

```python
x_plano_saude2_scaled
```

```python
y_plano_saude2_scaled
```

```python
# Kernel rbf com escala
regressor_svr_saude_rbf_scaled = SVR(kernel='rbf')
regressor_svr_saude_rbf_scaled.fit(x_plano_saude2_scaled, y_plano_saude2_scaled.ravel())
```

```python
grafico = px.scatter(x= x_plano_saude2_scaled.ravel(), y= y_plano_saude2_scaled.ravel())
grafico.add_scatter(x= x_plano_saude2_scaled.ravel(), y= regressor_svr_saude_rbf_scaled.predict(x_plano_saude2_scaled), name= 'Regressão')
grafico.show()
```

Base Preço das Casas - 83085.3090 - 73,39

```python
base_casas = pd.read_csv('../house_prices.csv')
x_casas = base_casas.iloc[:, 3:19].values
y_casas = base_casas.iloc[: ,2].values
x_casas_treinamento, x_casas_teste, y_casas_treinamento, y_casas_teste = train_test_split(x_casas, y_casas, test_size = 0.3, random_state = 0)
```

```python
x_casas_treinamento.shape
```

```python
x_casas_treinamento
```

```python
y_casas_treinamento
```

```python
x_casas_teste.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x_casas = StandardScaler()
x_casas_treinamento_scaler = scaler_x_casas.fit_transform(x_casas_treinamento)
scaler_y_casas = StandardScaler()
y_casas_treinamento_scaler = scaler_y_casas.fit_transform(y_casas_treinamento.reshape(-1,1))
```

```python
x_casas_treinamento_scaler
```

```python
y_casas_treinamento_scaler
```

```python
x_casas_teste_scaler = scaler_x_casas.fit_transform(x_casas_teste)
y_casas_teste_scaler = scaler_y_casas.fit_transform(y_casas_teste.reshape(-1,1))
```

```python
x_casas_teste_scaler
```

```python
y_casas_teste_scaler
```

```python
x_casas_treinamento_scaler.shape, y_casas_treinamento_scaler.shape
```

```python
x_casas_teste_scaler.shape, y_casas_teste_scaler.shape
```

```python
regressor_svr_casas = SVR(kernel='rbf')
regressor_svr_casas.fit(x_casas_treinamento_scaler, y_casas_treinamento_scaler.ravel())
```

```python
regressor_svr_casas.score(x_casas_treinamento_scaler, y_casas_treinamento_scaler)
```

```python
regressor_svr_casas.score(x_casas_teste_scaler, y_casas_teste_scaler)
```

```python
previsoes = regressor_svr_casas.predict(x_casas_teste_scaler)
previsoes
```

```python
y_casas_teste_scaler
```

```python
y_casas_teste_inverse = scaler_y_casas.inverse_transform(y_casas_teste_scaler)
previsoes_inverse = scaler_y_casas.inverse_transform(previsoes.reshape(-1, 1))
```

```python
y_casas_teste_inverse
```

```python
previsoes_inverse
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_casas_teste_inverse, previsoes_inverse)
```
